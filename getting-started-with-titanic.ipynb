## 3. Map Non-numericals to numericals
'Name', 'Sex', and 'Embarked' are strings. 
- 'Sex' can be mapped as: 0 = Male, 1 = Female.
- 'Embarked' can be mapped as: C = 0, Q = 1, S = 2.

```python
train_data['Sex'] = train_data['Sex'].map({'male':0, 'female':1})
train_data['Embarked'] = train_data['Embarked'].map({'C':0, 'Q':1, 'S':2})
```

```python
display(train_data.head())
display(train_data.loc[:, 'Sex'].value_counts(dropna=False))
display(train_data.loc[:, 'Embarked'].value_counts(dropna=False))
```

## 4. Feature Engineering
### Title feature 
'Name' contains the title of each individual such as 'Mr.', 'Mrs.', 'Dr.', etc. Therefore, we can extract this information into its own feature and drop the original 'Name' feature. Then map each value to a numerical.

```python
train_data['Title'] = train_data.Name.str.extract(' ([A-Za-z]+)\.', expand=False) # False means this will return a series instead of a dataframe.
train_data = train_data.drop(columns='Name')
train_data.head()
```

To map each value of Title to a numerical, we need to know all the possible values. Therefore, we can use a countplot.

```python
sns.catplot(data=train_data, y="Title", kind='count', height=10, aspect=2)
plt.show()
display(train_data.loc[:, 'Title'].value_counts(dropna=False))
```

- There are many titles that rarely occur compared to titles such as Mr, Mrs, Miss. So we can combine those rarely occurring classes into a single category. 
- Also, titles such as Miss and Ms are equivalent, so these will need to be combined into a single category.

```python
# Combine these rare categories into a single 'Others' category.
train_data['Title'] = train_data['Title'].replace(['Dr', 'Rev', 'Col', 'Major', 'Countess', 'Sir', 'Jonkheer', 'Lady', 'Capt', 'Don'], 'Others')
# Translate madame and madamosoille to Mrs and Miss because they are different names for the same thing.
train_data['Title'] = train_data['Title'].replace(['Mme'], 'Mrs')
train_data['Title'] = train_data['Title'].replace(['Mlle'], 'Miss')
train_data['Title'] = train_data['Title'].replace(['Ms'], 'Miss')

# Visualize new distribution.
sns.set(rc={'figure.figsize':(20, 10)})
fig, ax = plt.subplots(2, 1)

sns.countplot(data=train_data, y="Title", ax=ax[0])
sns.barplot(data=train_data, x="Title", y="Survived", ax=ax[1])

plt.show()
display(train_data.loc[:, 'Title'].value_counts(dropna=False))
```

Finally, with the finallized categories, we can map them to numericals.

```python
train_data['Title'] = train_data['Title'].map({'Master':0, 'Miss':1, 'Mr':2, 'Mrs':3, 'Others':4})
display(train_data.head())
display(train_data.loc[:, 'Title'].value_counts(dropna=False))
```

### FamilySize Feature
SibSp is the number of siblings and spouses of that individual aboard the Titanic. Parch is the number of parents and children of that individual aboard the Titanic. Therefore, we can determine the number of immediate family members of an individual aboard the Titanic using the equation:  
    
    FamilySize = 1 + SibSp + Parch

```python
train_data['FamilySize'] = 1 + train_data['SibSp'] + train_data['Parch']
display(train_data.head())
```

#### Finalized code
Here's the finallized code to impute 'Age'. Note that we include the rows with NaN in them to preserve as much of the original distribution as possible?

```python
train_data_median_age = train_data.median()['Age']
NaN_age_indexes = train_data['Age'][train_data['Age'].isnull()].index

for i in NaN_age_indexes:
    subset = train_data[
        (train_data['Pclass']==train_data.loc[i, 'Pclass']) &
        (train_data['SibSp']==train_data.loc[i, 'SibSp']) &
        (train_data['Parch']==train_data.loc[i, 'Parch']) &
        (train_data['Title']==train_data.loc[i, 'Title'])
    ]['Age']
    
    if subset.notnull().any(): # If the subset has values, then calc the subset median.
        median_age = subset.median() 
        train_data.at[i, 'Age'] = median_age
    else: # If the subset is null, that means no sample with the specified matching columns as the sample with the null 'Age' exists. Therefore, impute with the training data's median 'Age' value instead.
        train_data.at[i, 'Age'] = train_data_median_age
    
# sns.set(rc={'figure.figsize':(10, 10)})
fig, ax = plt.subplots(2, 2)

sns.boxplot(data=train_data, x="Pclass", y="Age", ax=ax[0][0])
sns.boxplot(data=train_data, x="SibSp", y="Age", ax=ax[0][1])
sns.boxplot(data=train_data, x="Parch", y="Age", ax=ax[1][0])
sns.boxplot(data=train_data, x="Title", y="Age", ax=ax[1][1])

plt.show()
```

Chi-squared correlation between Deck and Pclass returned a test statistic of 170 and a pvalue of 5.72e-29. Since the pvalue is much smaller than 0.05, we reject the null hypothesis that there is no relationship between Deck and Pclass and conclude that there is some relationship between Deck and Pclass.
- There is a 5.72e-29 probability that we observed a test statistic of 170 under the assumption that there is no relationship between Deck and Pclass. (i.e., very unlikely to have occured due to random chance).
- Also, the test statistic of 170 is much greater than the critical value of 23.685 which also supports the conclusion of rejecting the null hypothesis.

### Custom Dataset

Note that for __getitem__ returning a tuple instead of a dictionary simplifies the syntax when we get batches from the DataLoader.

```python
class PandasDataset(torch.utils.data.Dataset): # Make sure this custom dataset inherits Dataset.
    def __init__(self, dataframe):
        self.data = dataframe.values.astype(np.float32) # self.data is a numpy.ndarray
        self.ID = dataframe.iloc[:, 0].values.astype(np.float32)

    # Override
    def __getitem__(self, index):
        sample = torch.from_numpy(self.data[index]) # Get a row from data. This row contains the features and the label.
#         print(f"sample type: {type(sample)}")
#         print(f"type sample: {type(sample[1:-1])}")
#         print(f"type sample: {type(sample[1:-1])}")
#         x = torch.tensor(data=sample[1:-1])
        x = sample[1:-1]
#         print(f"x: {x}")
#         print(f"x type: {type(x)}")
#         print(f"x shape: {list(x.size())}")
#         y = torch.tensor(data=sample[-1])
        y = sample[-1]
#         print(f"y: {y}")
#         print(f"y type: {type(y)}")
#         print(f"y shape: {list(y.size())}")
#         print(f"y dim 0: {y.size(dim=0)}")
#         print(f"y dim 1: {y.size(dim=1)}")
#         print(f"y dim 2: {y.size(dim=2)}")
        sample_ID = self.ID[index]
        return (x, y, sample_ID)
#         return {'X': x, 'y': y}

    # Override
    def __len__(self):
        return len(self.data)

class TestPandasDataset(torch.utils.data.Dataset): # Make sure this custom dataset inherits Dataset.
    def __init__(self, dataframe):
        self.data = dataframe.values.astype(np.float32) # self.data is a numpy.ndarray
        self.ID = dataframe.iloc[:, 0].values.astype(np.float32)

    # Override
    def __getitem__(self, index):
        sample = torch.from_numpy(self.data[index])
        x = sample[1:]
        sample_ID = self.ID[index]
        return (x, sample_ID)

    # Override
    def __len__(self):
        return len(self.data)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    correct = 0
    epoch_loss = 0
    current = 0
    model.train() # Set the model to training mode.
    for batch, (X, y, ID) in enumerate(dataloader): # Get the batch num and the batch itself.
#         print(f"batch: {batch}, X: {X}, y: {y}, ID: {ID}")
#         print(f"type y: {type(y)}")
        # Compute prediction error.
        pred = model(X)
#         print(f"b4 pred type: {type(pred)}")
#         print(f"b4 pred shape: {list(pred.size())}")
        pred = torch.squeeze(pred)
#         print(f"  pred: {pred} | y: {y}")
#         print(f"  pred type: {type(pred)}")
#         print(f"  pred shape: {list(pred.size())}")
#         print(f"  y type: {type(y)}")
#         print(f"  y shape: {list(y.size())}")
        loss = loss_fn(pred, y)
        
        # Backprop.
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
#         print(f"type pred: {type(pred)}")
        t0 = torch.from_numpy(np.where(pred > 0.5, 1, 0))
#         print(f"t0: {t0}")
#         print(f"type t0: {type(t0)}")
#         print(f"y: {y}")
#         print(f"type y: {type(y)}")
        correct += (t0 == y).type(torch.float).sum().item()
        print(f"train correct: {correct}")
#         print(f"(t0 == y): {(t0 == y)}")
#         print(f"(t0 == y).type(torch.float): {(t0 == y).type(torch.float)}")
#         print(f"(t0 == y).type(torch.float).sum(): {(t0 == y).type(torch.float).sum()}")
#         print(f"(t0 == y).type(torch.float).sum().item(): {(t0 == y).type(torch.float).sum().item()}")
        # Show loss 
        loss, current = loss.item(), current + len(X)
        print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
            
    epoch_loss /= num_batches
    accuracy = (correct/size)*100
    print(f"accuracy: {accuracy}")
    history_dict['train loss'].append(epoch_loss)
    history_dict['train accuracy'].append(accuracy)
        
        
```

```python
def test(dataloader, model, loss_fn, optimizer):
    model.eval() # Set model to evaluation mode.
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for X, y, ID in dataloader:
            pred = model(X)
            pred = torch.squeeze(pred)
#             print(f" - pred: {pred} | y: {y}")
            test_loss += loss_fn(pred, y).item()
#             print(f" - np.argmax(pred, axis=1): {np.argmax(pred, axis=1)}")
#             print(f" - np.argmax(y, 1): {np.argmax(y, 1)}")
#             print(f" - (np.argmax(pred, 1) == np.argmax(y, 1)): {(np.argmax(pred, 1) == np.argmax(y, 1))}")
#             print(f" - (np.argmax(pred, 1) == np.argmax(y, 1)).type(torch.float).sum().item(): {(np.argmax(pred, 1) == np.argmax(y, 1)).type(torch.float).sum().item()}")
            
            t0 = torch.from_numpy(np.where(pred > 0.5, 1, 0))
            correct += (t0 == y).type(torch.float).sum().item()
            print(f"test correct: {correct}")
    
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
    
    history_dict['val loss'].append(test_loss)
    history_dict['val accuracy'].append(100*correct)
```

```python
def train_test_loop(epochs, dataloader, model, loss_fn, optimizer, scheduler, model_name, history_dict, save):
    # Loop thru epochs.
    for i in range(epochs):
        print(f"Epoch {i}\n-------------------------------")
        train(dataloader['train'], model, loss_fn, optimizer)
        test(dataloader['val'], model, loss_fn, optimizer)
        scheduler.step()
    plot_Acc_And_Loss(model_name, history_dict, save)
    print("DONE")
```

```python
import matplotlib.pyplot as plt 
import os

def plot_Acc_And_Loss(model_name, history_dict, save=True):
    """
    Plots loss and accuracy of train and val data over epochs.
    :return:
    """
    plt.plot(history_dict['train accuracy'])
    plt.plot(history_dict['val accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    if save: 
        if not os.path.exists(f'model_logs/{model_name}_logs/'):
            os.makedirs(f'model_logs/{model_name}_logs/')
        plt.savefig(f'model_logs/{model_name}_logs/{model_name}_accuracy.png')
    plt.show()

    plt.plot(history_dict['train loss'])
    plt.plot(history_dict['val loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    if save: 
        if not os.path.exists(f'model_logs/{model_name}_logs/'):
            os.makedirs(f'model_logs/{model_name}_logs/')
        plt.savefig(f'model_logs/{model_name}_logs/{model_name}_loss.png')
    plt.show()
```

```python
def predict(dataloader, model):
    model.eval()
    predictions = []
    with torch.no_grad():
        for X, ID in dataloader:
#             print(f"X: {X}")
            pred = model(X)
            pred = np.argmax(pred, axis=1)
#             print(f"pred: {pred.numpy().item()}")
            predictions.append(pred.numpy().item())
#             np.concatenate(list(map(lambda x: x[1].numpy(), pred)))
#             print(f"Predictions: {predictions}")
    
    return predictions
```

```python
class LogisticRegression(nn.Module):
    def __init__(self, input_shape):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(in_features=input_shape, out_features=1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        pred = self.layers(x)
        return pred

input_shape = len(dataset['train'][0][0])
lr_model = LogisticRegression(input_shape)
summary(lr_model, inputs_size=(batch_size, input_shape))
```
